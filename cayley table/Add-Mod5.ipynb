```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Cayley Table for Addition Modulo 5

S = [0, 1, 2, 3, 4]
print("+ |", " ".join(map(str, S)))
print("-" * 15)

for a in S:
    print(f"{a} |", end=" ")
    for b in S:
        result = (a + b) % 5
        print(result, end=" ")
    print()
```
