```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Cayley Table for Multiplication Modulo 4

S = [0, 1, 2, 3]
print("* |", " ".join(map(str, S)))
print("-" * 12)

for a in S:
    print(f"{a} |", end=" ")
    for b in S:
        result = (a * b) % 4
        print(result, end=" ")
    print()
```
