```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Incidence Matrix for an Undirected Graph

vertices = ['V1', 'V2', 'V3', 'V4']
edges = [('V1', 'V2'), ('V2', 'V3'), ('V3', 'V4'), ('V4', 'V1'), ('V1', 'V3')]

v_map = {v: i for i, v in enumerate(vertices)}
# Incidence matrix: Rows = Vertices, Columns = Edges
matrix = [[0 for _ in range(len(edges))] for _ in range(len(vertices))]

for e_idx, (u, v) in enumerate(edges):
    matrix[v_map[u]][e_idx] = 1
    matrix[v_map[v]][e_idx] = 1

print("Incidence Matrix (Rows: Vertices, Cols: Edges)")
print("   " + "  ".join([f"E{i+1}" for i in range(len(edges))]))
print("-" * 25)
for i, row in enumerate(matrix):
    print(f"{vertices[i]} | {'   '.join(map(str, row))}")
```
