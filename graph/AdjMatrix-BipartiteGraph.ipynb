```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Adjacency Matrix for a Bipartite Graph (K_2,2)

vertices = ['A', 'B', 'C', 'D']
# Set U = {A, B}, Set V = {C, D}
edges = [('A', 'C'), ('A', 'D'), ('B', 'C'), ('B', 'D')]

v_map = {v: i for i, v in enumerate(vertices)}
n = len(vertices)
matrix = [[0 for _ in range(n)] for _ in range(n)]

for u, v in edges:
    row, col = v_map[u], v_map[v]
    matrix[row][col] = 1
    matrix[col][row] = 1 # Undirected

print("   " + " ".join(vertices))
print("-" * 15)
for i, row in enumerate(matrix):
    print(f"{vertices[i]} | {' '.join(map(str, row))}")
```
