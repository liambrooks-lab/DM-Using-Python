```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Adjacency Matrix for a Directed Graph (with self-loop)

vertices = ['V1', 'V2', 'V3']
# Directed edges: V1->V2, V2->V3, V3->V1, and V1->V1 (self-loop)
edges = [('V1', 'V2'), ('V2', 'V3'), ('V3', 'V1'), ('V1', 'V1')]

v_map = {v: i for i, v in enumerate(vertices)}
n = len(vertices)
matrix = [[0 for _ in range(n)] for _ in range(n)]

for u, v in edges:
    row, col = v_map[u], v_map[v]
    matrix[row][col] = 1 # Only one direction!

print("   " + " ".join(vertices))
print("-" * 12)
for i, row in enumerate(matrix):
    print(f"{vertices[i]} | {' '.join(map(str, row))}")
```
