```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Verifying Handshaking Lemma on a Complete Graph (K4)

import itertools

vertices = ['1', '2', '3', '4']
# K4 has edges between ALL distinct pairs
edges = list(itertools.combinations(vertices, 2))

degree = {v: 0 for v in vertices}
for u, v in edges:
    degree[u] += 1
    degree[v] += 1

print("Graph: Complete Graph K4")
for v in vertices:
    print(f"deg({v}) = {degree[v]}")

total_deg = sum(degree.values())
print(f"\nSum of degrees: {total_deg}")
print(f"2 x Number of edges ({len(edges)}): {2 * len(edges)}")

if total_deg == 2 * len(edges):
    print("-> Handshaking Lemma VERIFIED for Complete Graph!")
```
