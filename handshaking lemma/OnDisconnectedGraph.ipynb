```python
# Name: Rudranarayan Jena  PRN: 20250802081
# Topic: Verifying Handshaking Lemma on a Disconnected Graph

vertices = ['A', 'B', 'C', 'D', 'E']
# Component 1: A-B-C, Component 2: D-E (No link between components)
edges = [('A', 'B'), ('B', 'C'), ('D', 'E')]

degree = {v: 0 for v in vertices}
for u, v in edges:
    degree[u] += 1
    degree[v] += 1

print("Graph: Disconnected Graph")
for v in vertices:
    print(f"deg({v}) = {degree[v]}")

sum_deg = sum(degree.values())
print(f"\nSum of all degrees = {sum_deg}")
print(f"2 x Number of edges = {2 * len(edges)}")

if sum_deg == 2 * len(edges):
    print("-> Handshaking Lemma verified even for Disconnected Graphs!")
```
